```python
# Cleaning the Hill/Flat column 
df['Hill/Flat'] = df['Hill/Flat'].str.replace(' ', '').str.lower()
df['Hill/Flat'] = df['Hill/Flat'].str.replace('glat', 'flat').str.lower()
df['Hill/Flat'].loc[df['Hill/Flat'].str.startswith('h', na=False)] = 'hill'
df['Hill/Flat'].loc[df['Hill/Flat'].str.startswith('f', na=False)] = 'flat'
df['Hill/Flat'].unique()

# Combine the Time(sec) and Blocktime columns 
df.loc[pd.isnull(df['Time(Sec)']), 'Time(Sec)'] = df['Blocktime']

# Make the df.columns printout look nicer 
df.columns = [str(i) for i in df.columns]

# Make sure all route numbers have 3 digits -- not sure why this was needed though
df['Route'] = df['Route'].astype(str)
df['Route'] = df['Route'].apply(lambda x: x.zfill(3))

# Make the column names nicer -- remove the spaces 
df = df.rename(columns={'1 yrd': '1yd', '1.5 yrd': '1.5yd', '2 yrd': '2yd', '3 yrd':'3yd',
                        '4 yrd': '4yd', '5 yrd': '5yd', '6 yrd': '6yd'})

# The 1yrd column was object (according to df.info()) and needed to be converted to int 
df['1yd'] = pd.to_numeric(df['1yd'], errors='coerce')
```

```python
# Created a new df that only included the size columns and time --> 
# Take the mean and count for each bin combination in the dataframe
df_group = df.groupby(['Inside/Curb', '16', '20', '32', '64', '96', '1yd', '1.5yd', '2yd', '3yd', '4yd', '5yd', '6yd'])[
        'Time(Sec)'].agg(['mean','count']).reset_index()

df_group.sort_values(['count'], ascending=False) 
# Most common combinations: 1 96-container (1323),  
```

```python
df_group.loc[df_group['count']==0]
```

```python
large_count = 0
df_group = df_group.sort_values(['Inside/Curb'])
df_group = df_group.rename(columns={'mean': 'Time(Sec)'})


# Convert data to numeric type in these columns
vol_list = ['16', '20', '32', '64', '96', '1yd', '1.5yd', '2yd', '3yd', '4yd', '5yd', '6yd']
for vol in vol_list:
    df_group[vol] = pd.to_numeric(df_group[vol], errors='coerce')
    df_group = df_group.dropna()
#     df_group_small = df_group[df_group['yard'] == 'N']
#     df_group_large = df_group[df_group['yard'] == 'Y']

#Split the dataset into Inside and Curb

df_c = df_group[(df_group['Inside/Curb'] == 'C')]
df_i = df_group[(df_group['Inside/Curb'] == 'I')]
df_ic = df_group[(df_group['Inside/Curb'] == 'IC')]

# Create two dataframes Inside and Curb for modelling with count>=0 - not sure the reason for this ? 

df_c_large = df_group[(df_group['Inside/Curb'] == 'C') & (df_group['count'] >= large_count)]
df_i_large = df_group[(df_group['Inside/Curb'] == 'I') & (df_group['count'] >= large_count)]

# Create a list of features

x_list = ['16', '20', '32', '64', '96', '1yd', '1.5yd', '2yd', '3yd', '4yd', '5yd', '6yd']

# Create input dataframe for regression
X1, y1 = df_c_large[x_list], df_c_large['Time(Sec)']
X2, y2 = df_i_large[x_list], df_i_large['Time(Sec)']

#Linear regression performed on training dataframes
reg1 = LinearRegression().fit(X1, y1)  # Regression for curb 
reg2 = LinearRegression().fit(X2, y2)  # Regression for inside 
reg1_pred = reg1.predict(df_c[x_list]) # Predicted time for curb
reg2_pred = reg2.predict(df_i[x_list]) # Predicted time for inside 

#Add the pred column to the original dataframe for inside and curb
df_c['pred'] = reg1_pred
df_i['pred'] = reg2_pred

# Why do this??
df_c.loc[df_c['count'] < large_count, 'Time(Sec)'] = df_c['pred']
df_i.loc[df_i['count'] < large_count, 'Time(Sec)'] = df_i['pred']

# Append all dataframes together 
df_all = pd.concat([df_c, df_i])
df_all = pd.concat([df_all, df_ic])
df_all

# # Subsetting the data frame
# df_all = df_all[['Inside/Curb'] + vol_list + ['Time(Sec)']]
# vol_list = ['16', '20', '32', '64', '96', '1yd', '1.5yd', '2yd', '3yd', '4yd', '5yd', '6yd']

# # Add a new column that shows how many containers are in each row 
# df_all['sum'] = df_all[vol_list].sum(axis=1)
# df_all = df_all[df_all['sum'] > 0] # Exclude the weird rows where there are no containers 
```
