# This notebook creates the matrix that gives an overview of service time estimates 

```
import pandas as pd
```

```
df_pred = pd.read_excel('Final_Predictions_Routes975_998-990.xlsx')
df_pred.groupby('32 gal')['time_pred'].median()
```

```
from collections import defaultdict
```

```
df_16 = df_pred.loc[(df_pred['16 gal']!= 0) & (df_pred['20 gal']== 0) & (df_pred['32 gal']== 0) & (df_pred['64 gal']== 0) & (df_pred['96 gal']== 0)]
df_20 = df_pred.loc[(df_pred['16 gal']== 0) & (df_pred['20 gal']!= 0) & (df_pred['32 gal']== 0) & (df_pred['64 gal']== 0) & (df_pred['96 gal']== 0)]


df_32 = df_pred.loc[(df_pred['16 gal']== 0) & (df_pred['20 gal']== 0) & (df_pred['32 gal']!= 0) & (df_pred['64 gal']== 0) & (df_pred['96 gal']== 0)]
df_64 = df_pred.loc[(df_pred['16 gal']== 0) & (df_pred['20 gal']== 0) & (df_pred['32 gal']== 0) & (df_pred['64 gal']!= 0) & (df_pred['96 gal']== 0)]
df_96 = df_pred.loc[(df_pred['16 gal']== 0) & (df_pred['20 gal']== 0) & (df_pred['32 gal']== 0) & (df_pred['64 gal']== 0) & (df_pred['96 gal']!= 0)]

dfs = [df_16, df_20, df_32, df_64, df_96]
gals = [16,20,32,64,96]

# How to get my for loop working????? SAD DAY 
# for df, gal in zip(dfs,gals):
#     if len(df)>0:
#         df = df.groupby(['Neighborhood',f'{gal} gal', 'Has Key'], as_index=False)['time_pred'].median().round(2)
#         df['Bin Size'] = f'{gal} gal'
#         df.rename(columns={f'{gal} gal':'# Units'}, inplace=True)
#         print(df)
        
if len(df_16)>0:
    df_16 = df_20.groupby(['Neighborhood','16 gal', 'Has Key'], as_index=False)['time_pred'].median().round(2)
    df_16['Bin Size'] = '16 gal'
    df_16.rename(columns={'16 gal':'# Units'}, inplace=True)

if len(df_20)>0:
    df_20 = df_20.groupby(['Neighborhood','20 gal', 'Has Key'], as_index=False)['time_pred'].median().round(2)
    df_20['Bin Size'] = '20 gal'
    df_20.rename(columns={'20 gal':'# Units'}, inplace=True)

if len(df_32)>0:
    df_32 = df_32.groupby(['Neighborhood','32 gal', 'Has Key'], as_index=False)['time_pred'].median().round(2)
    df_32['Bin Size'] = '32 gal'
    df_32.rename(columns={'32 gal':'# Units'}, inplace=True)

if len(df_64)>0:
    df_64 = df_64.groupby(['Neighborhood','64 gal', 'Has Key'], as_index=False)['time_pred'].median().round(2)
    df_64['Bin Size'] = '64 gal'
    df_64.rename(columns={'64 gal':'# Units'}, inplace=True)


if len(df_96)>0:
    df_96 = df_96.groupby(['Neighborhood','96 gal', 'Has Key'], as_index=False)['time_pred'].median().round(2)
    df_96['Bin Size'] = '96 gal'
    df_96.rename(columns={'96 gal':'# Units'}, inplace=True)

# Figure out how to not hard code this
final_matrix = df_32.append([df_64, df_96])
final_matrix = final_matrix[['Neighborhood', 'Bin Size', '# Units', "Has Key", 'time_pred']]
final_matrix = final_matrix.loc[final_matrix['# Units']< 5]
```

```
final_matrix.to_excel('final_matrix.xlsx', index=False)
```
