# 3. Create subset dfs that include only a single 32, 64, and 96 

```python
def subset_16gal(df):
    return df.loc[(collated[16]==1) & (df[20]==0) & (df[32]==0) & (df[64]==0) & (df[96]==0)] 

def subset_20gal(df):
    return df.loc[(collated[16]==0) & (df[20]==1) & (df[32]==0) & (df[64]==0) & (df[96]==0)] 

def subset_32gal(df):
    # return df.loc[(collated['16 gal']==0) & (df['20 gal']==0) & (df[32]==1) & (df['64 gal']==0) & (df['96 gal']==0)]
    return df.loc[(collated[16]==0) & (df[20]==0) & (df[32]==1) & (df[64]==0) & (df[96]==0)] 

def subset_64gal(df):
    #return df.loc[(collated['16 gal']==0) & (df['20 gal']==0) & (df['32 gal']==0) & (df['64 gal']==1) & (df['96 gal']==0)] 
    return df.loc[(collated[16]==0) & (df[20]==0) & (df[32]==0) & (df[64]==1) & (df[96]==0)] 

def subset_96gal(df):
    #return df.loc[(collated['16 gal']==0) & (df['20 gal']==0) & (df['32 gal']==0) & (df['64 gal']==0) & (df['96 gal']==1)]  
    return df.loc[(collated[16]==0) & (df[20]==0) & (df[32]==0) & (df[64]==0) & (df[96]==1)] 
```

```python
rgg_32 = subset_32gal(rgg)

rgg_64 = subset_64gal(rgg) 

rgg_96 = subset_96gal(rgg)
```

```python
rss_32 = subset_32gal(rss)

rss_64 = subset_64gal(rss) 

rss_96 = subset_96gal(rss)
```

```python
collated_16 = subset_16gal(collated)
collated_20 = subset_20gal(collated)
collated_32 = subset_32gal(collated)
collated_64 = subset_64gal(collated)
collated_96 = subset_96gal(collated)
```

```python
def create_comparison_boxplot(df1,df2,df3, company):
    orange = '#ffab03'
    teal = "#48bec2"
    colors = [teal, orange]

    fig, (ax1,ax2, ax3) = plt.subplots(nrows = 1, ncols = 3, figsize=(10,8))

    sns.boxplot(x = '32 gal', y = "Time(Sec)", data = df1, palette = colors, ax=ax1)\
                .set_title("32 gal bin")

    sns.boxplot(x = '64 gal', y = "Time(Sec)", data = df2, palette = colors, ax=ax2)\
                .set_title("64 gal bin")

    sns.boxplot(x = '96 gal', y = "Time(Sec)", data = df3, palette = colors, ax=ax3)\
                .set_title("96 gal bin")

    title = fig.suptitle(f"{company} Service Times for A Single 32, 64, and 96 Bin", fontsize=18, fontweight = 'bold')
```

##  Create Inside/Outside Times based on #units

```python
collated.info()
```

```python
collated['Inside/Curb'].unique()
```

```python
df_units= collated.copy()
#df_units['Inside/Curb'].unique()
df_units['units']=df_units[16]+df_units[20]+df_units[32]+df_units[64]+df_units[96]
df_units['Inside/Curb']=df_units['Inside/Curb'].replace('I','i')
df_units['Inside/Curb']=df_units['Inside/Curb'].replace('i','Inside')
df_units['Inside/Curb']=df_units['Inside/Curb'].replace('C','Curbside')
df_units=df_units.loc[df_units['Inside/Curb'].isin(['Inside','Curbside'])]
df_summary=df_units.groupby(['Company','Inside/Curb','units'])['Time(Sec)'].median().to_frame(name='median').reset_index()
df_summary['units_bin']=np.where(df_summary['units']>=5,'Greater than 5',df_summary['units'])
df_summary['key']= df_summary.agg('{0[Company]} with {0[Inside/Curb]}'.format, axis=1)
df_summary=df_summary.loc[df_summary['units']>0]
df_summary
```

```python
orange = '#e66b2e'
teal = "#48bec2"
red="#bf1708"
blue="#0870bf"

colors = [orange,teal,red,blue]
fig, (ax1,ax2, ax3, ax4, ax5) = plt.subplots(nrows = 1, ncols = 5, sharey=True,figsize=(16,8))
ax = [ax1,ax2,ax3,ax4,ax5]
units=df_summary['units_bin'].unique()
for a,u in zip(ax,units):
    sns.barplot(x = 'units_bin', y = "median", data = df_summary.loc[df_summary['units_bin']==u], palette = colors, hue='key', estimator=np.median, ax=a)
    a.set_xlabel('Number of Bins')
    a.set_ylabel('Median Service Time')

```
