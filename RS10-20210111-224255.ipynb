# Merge the time estimation and RS10 data

```python
def estimate(df_group, df_rs, df_ac):
    df_rs_key = df_rs.merge(df_ac, on='Z1SVC#', how='left')
    df_rs_key['Inside/Curb'] = df_rs_key['Inside/Curb'].fillna('C')

    bin_list = ['16', '20', '32', '64', '96', '1yd', '1.5yd', '2yd', '3yd', '4yd', '5yd', '6yd']
    '''
    To convert the volumn into columns so that they can be joined with the time estimation
    '''
    df_rs_key['r_size'] = df_rs_key['Z1DESC'].str.split().str[0].str.lower()
    yd_pivot = df_rs_key.pivot_table(values='Z1RUNT',
                index=[df_rs_key.index, 'Z1AID'],
                     columns='r_size').reset_index().drop(columns=['level_0'], axis=1)
    r_pivot = df_rs_key.pivot_table(values='Z1RUNT',
                index=[df_rs_key.index, 'Z1AID'],
                     columns='Z1RSIZ').reset_index().drop(columns=['level_0'], axis=1)
    g_pivot = df_rs_key.pivot_table(values='Z1UNIT',
                    index=[df_rs_key.index, 'Z1AID'],
                         columns='Z1SIZE').reset_index().drop(columns=['level_0'], axis=1)
    o_pivot = df_rs_key.pivot_table(values='Z1OUNT',
                    index=[df_rs_key.index, 'Z1AID'],
                         columns='Z1OSIZ').reset_index().drop(columns=['level_0'], axis=1)
    pivot_all = yd_pivot
    pivot_all = pd.concat([pivot_all, yd_pivot])
    pivot_all = pd.concat([pivot_all, g_pivot])
    pivot_all = pd.concat([pivot_all, o_pivot])
    pivot_all = pivot_all.groupby(['Z1AID']).sum().reset_index()
    df_rs_pivot = df_rs_key.merge(pivot_all, on = 'Z1AID', how='left')

    df_rs_pivot.columns = [str(i) for i in df_rs_pivot.columns]
    df_rs_pivot = df_rs_pivot.fillna(0)#.drop(columns=['0'], axis=1)
    df_rs_pivot = df_rs_pivot.rename(columns={'200': '1 yrd'})
    bin_in = [i for i in df_rs_pivot.columns if i in bin_list]
    df_rs_pivot_sum = df_rs_pivot.groupby([i for i in df_rs_pivot.columns if i not in bin_in
                                          ])[bin_in].sum().reset_index()
    for bin_vol in bin_list:
        if bin_vol not in df_rs_pivot_sum.columns:
            df_rs_pivot_sum[bin_vol] = 0.0

    df_rs_sum = df_rs_pivot_sum.merge(df_group,
                            on=df_group.drop(columns=['Time(Sec)'], axis=1).columns.tolist(), how='left')
    #df_final = extrapolate(df_rs_sum, df_group)
    df_final = df_rs_sum
    return df_final
```

```python
df_570 = df_rs[df_rs['Z1ROUT'].isin(['570', '571', '572', '573'  '574', '575'])]


df_res_570 = estimate(df_group, df_570, df_ac)
df_res_570['Time(Sec)'] = df_res_570['Time(Sec)'].fillna(df_res_570['Time(Sec)'].mean())
```
